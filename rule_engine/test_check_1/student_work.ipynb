```python
import networkx as nx
import random
```

```python
# Создаём ER-граф
n = 500
p = 0.016
G_er = nx.erdos_renyi_graph(n, p, seed=42)
print(f"Создан граф с {G_er.number_of_nodes()} вершинами")
```

```python
# Считаем метрики
avg_degree = sum(dict(G_er.degree()).values()) / n
clustering = nx.average_clustering(G_er)
metrics_er = {
    'avg_degree': avg_degree,
    'clustering': clustering
}
print(f"Метрики: {metrics_er}")
```
